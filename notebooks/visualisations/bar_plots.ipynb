```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('../../cleaned_fighters_data.csv')

# Define success threshold and add a binary column for high and low success
success_threshold = 70
df['high_success'] = (df['win_percentage'] >= success_threshold).astype(int)

# List of metrics for bar plots
metrics = ['strikes_per_min', 'striking_accuracy', 'strikes_absorbed_per_min', 
           'striking_defense', 'takedown_avg', 'takedown_accuracy', 'takedown_defense', 'submission_avg']

# Bar Plot - Average Metrics for High vs Low Success Fighters
plt.figure(figsize=(15, 8))
for i, metric in enumerate(metrics, 1):
    plt.subplot(2, 4, i)
    sns.barplot(data=df, x='high_success', y=metric, errorbar=None)
    # Keep a descriptive title without being too long
    plt.title(f"Average {metric.replace('_', ' ').title()}")  # Keeping titles more descriptive
    plt.xlabel("High Success (1 = Yes, 0 = No)")
    plt.ylabel(f"Average {metric.replace('_', ' ').title()}")

# Increase the space between subplots to avoid overlapping titles
plt.tight_layout(pad=5.0)
plt.show()
```
