```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('../../cleaned_fighters_data.csv')

# List of metrics to plot against win percentage
metrics = ['strikes_per_min', 'striking_accuracy', 'strikes_absorbed_per_min', 
           'striking_defense', 'takedown_avg', 'takedown_accuracy', 'takedown_defense', 'submission_avg']

# Number of rows and columns for subplots
n_cols = 4  # 4 columns in the grid
n_rows = (len(metrics) + n_cols - 1) // n_cols  # Compute number of rows to fit all plots

# Create a figure with subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, 8))

# Flatten the axes array to make it easier to iterate through
axes = axes.flatten()

# Scatter Plot - Metrics vs. Win Percentage with Line of Best Fit
for i, metric in enumerate(metrics):
    sns.regplot(data=df, x=metric, y='win_percentage', scatter_kws={'s': 100}, line_kws={'color': 'red'}, ax=axes[i])
    axes[i].set_title(f"{metric.replace('_', ' ').title()} vs Win Percentage")
    axes[i].set_xlabel(metric.replace('_', ' ').title())
    axes[i].set_ylabel("Win Percentage")

# Remove empty subplots if any
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust the layout for better spacing
plt.tight_layout()
plt.show()
```
