```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('../../cleaned_fighters_data.csv')

# List of metrics to plot against win percentage
metrics = ['strikes_per_min', 'striking_accuracy', 'strikes_absorbed_per_min', 
           'striking_defense', 'takedown_avg', 'takedown_accuracy', 'takedown_defense', 'submission_avg']

# Scatter Plot - Metrics vs. Win Percentage with Line of Best Fit
for metric in metrics:
    plt.figure(figsize=(8, 6))
    sns.regplot(data=df, x=metric, y='win_percentage', scatter_kws={'s': 100}, line_kws={'color': 'red'})
    plt.title(f"{metric.replace('_', ' ').title()} vs Win Percentage with Line of Best Fit")
    plt.xlabel(metric.replace('_', ' ').title())
    plt.ylabel("Win Percentage")
    plt.show()

```
